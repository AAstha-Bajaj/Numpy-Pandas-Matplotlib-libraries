```python
df = pd.read_csv('record.csv')
df
```

```python
plt.bar(df['batsman'], df['2015'], width = 0.2)
```

```python
plt.bar(np.arange(df.shape[0])-0.3, df['2015'], width = 0.3)
plt.bar(np.arange(df.shape[0]), df['2016'], width = 0.3)
plt.bar(np.arange(df.shape[0])+0.3, df['2017'], width = 0.3)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
#a problem
children = [10,43,35,66,23]
colors = ['red redredredredredred', 'blueblueblueblueblueblue','orangeorangeorangeorange','yellowyellowyellow', 'pinkpinkpink']
plt.bar(colors, children, color='black')
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.bar(df['batsman'], df['2017'], label='2017')
plt.bar(df['batsman'], df['2016'], bottom=df['2017'], label='2016')
plt.bar(df['batsman'], df['2015'], bottom=(df['2016'] + df['2017']), label='2015')

plt.legend()
plt.show()
```

**Series from Lists**

```python
#string 
country = ['India', 'Pakistan', 'USA', 'Nepal', 'Srilanka']

pd.Series(country)
```

```python
#integers
runs =[12,45,77,43]

pd.Series(runs)
```

```python
#custom index
marks = [23,4,64,32]
sub = ['eng','hind','maths','sceince']

pd.Series(marks, index=sub)
```

```python
#setting a name
marks = pd.Series(marks, index = sub, name = "raj's marks")
marks
```

**Series from dict**

```python
marks = {
    'eng': 21,
    'hindi': 32,
    'mths': 32,
    'science':56
}

marks_series = pd.Series(marks, name = "raj's marks")
marks_series
```
